```python
columnas_binarias = [
    'DIABETES', 'EPOC', 'ASMA', 'INMUSUPR', 'HIPERTENSION', 'OTRA_COM',
    'CARDIOVASCULAR', 'OBESIDAD', 'RENAL_CRONICA', 'TABAQUISMO', 
    'INTUBADO', 'NEUMONIA', 'UCI', 'EMBARAZO'
]

print("Iniciando la limpieza de códigos '98' a NaN...")

for df in [df1, df2]:
    for col in columnas_binarias:
        if col in df.columns:
            df[col] = df[col].replace({98: np.nan}) 
            df[col] = pd.to_numeric(df[col], errors='coerce') 

print("Limpieza de códigos 98 completada.")
```

```python
columnas_df1 = df1.columns.tolist()

df2_alineado = df2.reindex(columns=columnas_df1) 

df_unificado = pd.concat([df1, df2_alineado], ignore_index=True)

print("\n--- Resultado de la Concatenación ---")
print(f"Número de registros en df1: {len(df1)}")
print(f"Número de registros en df2: {len(df2)}")
print(f"Total de registros en df_unificado: {len(df_unificado)}")
print(f"Columnas del DataFrame unificado: {df_unificado.shape[1]}")
print(df_unificado.head())
```

```python
df_unificado['FECHA_DEF'] = pd.to_datetime(df_unificado['FECHA_DEF'], errors='coerce')

df_unificado['FALLECIO'] = df_unificado['FECHA_DEF'].apply(lambda x: 1 if pd.notna(x) else 0)

print("\nConteo de fallecimientos:")
print(df_unificado['FALLECIO'].value_counts())
```

```python
df_unificado['EDAD'] = pd.to_numeric(df_unificado['EDAD'], errors='coerce')

bins_edad = [0, 18, 36, 60, df_unificado['EDAD'].max() + 1] 
labels_edad = ['0-17', '18-35', '36-59', '60+']

df_unificado['GRUPO_EDAD'] = pd.cut(
    df_unificado['EDAD'], 
    bins=bins_edad, 
    labels=labels_edad, 
    right=False, 
    include_lowest=True
)

print("\nConteo por grupo de edad:")
print(df_unificado['GRUPO_EDAD'].value_counts(dropna=False))
```

```python
cfr_por_estado = df_unificado.groupby('ENTIDAD_RES')['FALLECIO'].mean().mul(100).sort_values(ascending=False).reset_index()
cfr_por_estado.columns = ['ENTIDAD_RES', 'CFR (%)']

mapa_entidades = {
    1: 'AGUASCALIENTES', 2: 'BAJA CALIFORNIA', 3: 'BAJA CALIFORNIA SUR',
    4: 'CAMPECHE', 5: 'COAHUILA DE ZARAGOZA', 6: 'COLIMA',
    7: 'CHIAPAS', 8: 'CHIHUAHUA', 9: 'CIUDAD DE MÉXICO',
    10: 'DURANGO', 11: 'GUANAJUATO', 12: 'GUERRERO',
    13: 'HIDALGO', 14: 'JALISCO', 15: 'MÉXICO',
    16: 'MICHOACÁN DE OCAMPO', 17: 'MORELOS', 18: 'NAYARIT',
    19: 'NUEVO LEÓN', 20: 'OAXACA', 21: 'PUEBLA',
    22: 'QUERÉTARO', 23: 'QUINTANA ROO', 24: 'SAN LUIS POTOSÍ',
    25: 'SINALOA', 26: 'SONORA', 27: 'TABASCO',
    28: 'TAMAULIPAS', 29: 'TLAXCALA', 30: 'VERACRUZ DE IGNACIO DE LA LLAVE',
    31: 'YUCATÁN', 32: 'ZACATECAS', 99: 'NO ESPECIFICADO' 
}

cfr_por_estado['ENTIDAD_RES'] = pd.to_numeric(cfr_por_estado['ENTIDAD_RES'], errors='coerce', downcast='integer')

cfr_por_estado['Entidad Federativa'] = cfr_por_estado['ENTIDAD_RES'].map(mapa_entidades)

cfr_por_estado = cfr_por_estado[['Entidad Federativa', 'CFR (%)']]

print("Tasa de Fatalidad por Caso (CFR) de Todas las Entidades")
print(cfr_por_estado.round(2))
```

```python
# --- Paso 1: Asegurar el formato de fecha ---
# Convertir las columnas de fecha a tipo datetime. 'coerce' convierte fechas inválidas a NaT.
df_unificado['FECHA_SINTOMAS'] = pd.to_datetime(df_unificado['FECHA_SINTOMAS'], errors='coerce')
df_unificado['FECHA_DEF'] = pd.to_datetime(df_unificado['FECHA_DEF'], errors='coerce')

# --- Paso 2: Filtrar y Contar Casos Confirmados ---
# Se consideran confirmados los códigos 1, 2 y 3 en CLASIFICACION_FINAL_COVID.
casos_confirmados = df_unificado[df_unificado['CLASIFICACION_FINAL_COVID'].isin([1, 2, 3])].copy()

# A. Casos Confirmados por Día (usando la fecha de síntomas)
serie_casos = casos_confirmados.groupby('FECHA_SINTOMAS').size().reset_index(name='Casos_Confirmados')
serie_casos.rename(columns={'FECHA_SINTOMAS': 'Fecha'}, inplace=True)

# B. Defunciones por Día (usando la fecha de defunción)
# Filtramos solo los registros con FECHA_DEF válida (no nula)
df_defunciones = casos_confirmados[pd.notna(casos_confirmados['FECHA_DEF'])].copy()
serie_defunciones = df_defunciones.groupby('FECHA_DEF').size().reset_index(name='Defunciones')
serie_defunciones.rename(columns={'FECHA_DEF': 'Fecha'}, inplace=True)

# --- Paso 3: Unir las dos series en una sola tabla ---

# Fusionar Casos y Defunciones usando la 'Fecha' (Outer Join para incluir días con un solo evento)
serie_tiempo_final = pd.merge(
    serie_casos,
    serie_defunciones,
    on='Fecha',
    how='outer'
).sort_values('Fecha')

# Rellenar los días sin registro (NaN) con 0
serie_tiempo_final = serie_tiempo_final.fillna(0)

# Asegurar que las columnas de conteo sean de tipo entero
serie_tiempo_final['Casos_Confirmados'] = serie_tiempo_final['Casos_Confirmados'].astype(int)
serie_tiempo_final['Defunciones'] = serie_tiempo_final['Defunciones'].astype(int)

# --- PASO 4: CALCULAR LA TASA DE FATALIDAD DIARIA (CFR) ---

# Para evitar la división por cero (si hay 0 casos confirmados un día)
serie_tiempo_final['CFR_Diario (%)'] = np.where(
    serie_tiempo_final['Casos_Confirmados'] > 0,
    (serie_tiempo_final['Defunciones'] / serie_tiempo_final['Casos_Confirmados']) * 100,
    0.0 # Si no hay casos, el CFR es 0 (o NaN, pero 0 es más limpio para visualización)
)

print("--- Vista Previa de la Tabla de Serie de Tiempo (CON CFR Diario) ---")
# Usamos .tail() para ver los últimos días y el nuevo porcentaje
print(serie_tiempo_final.tail().round(2))
```

```python
# Lista de las columnas de comorbilidad a contar
columnas_comorbilidad = [
    'DIABETES', 'EPOC', 'ASMA', 'INMUSUPR', 'HIPERTENSION', 'OTRA_COM', 
    'CARDIOVASCULAR', 'OBESIDAD', 'RENAL_CRONICA', 'TABAQUISMO'
]

# Crear una copia de las columnas de interés para evitar SettingWithCopyWarning
df_conteo = df_unificado[['FALLECIO'] + columnas_comorbilidad].copy()

# Contar cuántas veces aparece el valor '1' (Sí) en cada fila. 
# La función .sum(axis=1) suma los valores en cada fila.
# Ignora los valores 2 y NaN.
df_conteo['NUM_COMORBILIDADES'] = df_conteo[columnas_comorbilidad].eq(1).sum(axis=1)

# Asignar el conteo de comorbilidades al DataFrame unificado principal (opcional, pero útil)
df_unificado['NUM_COMORBILIDADES'] = df_conteo['NUM_COMORBILIDADES']

# Definir los límites de los grupos de comorbilidad
# Usamos 3.5 como límite para incluir todos los valores de 3 en adelante
bins_comor = [-0.5, 0.5, 1.5, 2.5, df_unificado['NUM_COMORBILIDADES'].max() + 1]
labels_comor = ['0 comorbilidades', '1 comorbilidad', '2 comorbilidades', '3+ comorbilidades']

# Crear la columna categórica
df_unificado['GRUPO_COMORBILIDADES'] = pd.cut(
    df_unificado['NUM_COMORBILIDADES'],
    bins=bins_comor,
    labels=labels_comor,
    right=False,  # El intervalo incluye el límite inferior
    include_lowest=True
)

# Calcular la Tasa de Fatalidad (CFR)
cfr_por_comor = df_unificado.groupby('GRUPO_COMORBILIDADES')['FALLECIO'].mean().mul(100).reset_index(name='CFR (%)')

# Renombrar la columna del grupo
cfr_por_comor.columns = ['Grupo Comorbilidad', 'CFR (%)']

print("Tasa de Fatalidad por Número de Comorbilidade")
print(cfr_por_comor.round(2))
```

```python
nombre_carpeta = 'DataProcessing_Proyect'
nombre_archivo_ts = 'defuncionas_por_comorbilidades.csv'
ruta_completa_ts = os.path.join(nombre_carpeta, nombre_archivo_ts)

if not os.path.exists(nombre_carpeta):
    os.makedirs(nombre_carpeta)

cfr_por_comor.to_csv(
    ruta_completa_ts,
    index=False,
    encoding='utf-8' 
)

print(f"Ha sido exportada a: '{ruta_completa_ts}'")
```
